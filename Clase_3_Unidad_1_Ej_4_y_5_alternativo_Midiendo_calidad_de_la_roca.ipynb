## Calidad de rocas

**Indice RQD**

El índice RQD (del inglés Rock Quality Designation) es un parámetro geotécnico que se utiliza para evaluar la calidad de la roca en base al grado de fracturamiento observado en los testigos de perforación (testigos de roca extraídos mediante perforación con testigo continuo).

**¿Qué es?**

El RQD representa el porcentaje de la longitud total de testigo recuperado que está formado por fragmentos de roca mayores de 10 cm de longitud, medidos a lo largo del eje del testigo, excluyendo los fragmentos más pequeños o triturados. Se expresa en porcentaje y fue introducido por Don Deere en 1964.

**¿Cómo se calcula?**


$$RQD=\frac{\mathrm{Longitud~total~de~fragmentos~mayores~a~10cm}}{\mathrm{Longitud~total~de~perforación}}*100$$
